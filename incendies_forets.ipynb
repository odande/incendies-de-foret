```python
#import du fichier excel sur les incendies

import pandas as pd

df=pd.read_excel('incendies_1.xlsx')
```

```python
#Création de la fonction avec l'API du gouvernement français

def API_adresse(coordonnees):
  
  coordonnees=coordonnees.replace(' ','+')
  link_main = "https://api-adresse.data.gouv.fr/search/?q="
  link = link_main + coordonnees
  import requests

  r = requests.get(link).json()

  point = r['features'][0]['geometry']['coordinates'][::-1]

  
  return point
```

```python
#test de la fonction

API_adresse("Cagnano")
```

```python
#Travail sur la base afin de créer une base strictement nécéssaire a ce que je souhaoite afficher sur la carte

import pandas as pd

inc_foret= pd.DataFrame(df, columns = ['Annee', "Nom de la commune"])


inc_foret["Nom de la commune"].info()

#convertion de la colonne Nom de la commune en chaine de caractère
inc_foret['coordonnees']=inc_foret["Nom de la commune"].convert_dtypes('string')

inc_foret['coordonnees'].info()


#Application nde la foction précédente avec un apply
inc_foret['coordonnees1']=inc_foret['coordonnees'].apply(API_adresse)

```

```python
#Création de l'affichage géocodée

import folium

m = folium.Map(location=inc_foret['coordonnees1'][0],zoom_start=5)

for index, ligne in inc_foret.iterrows():
  folium.Marker(
     location=ligne['coordonnees1'],
     popup=ligne["Nom de la commune"],
     icon= folium.Icon(color = 'red')).add_to(m)
m
```
